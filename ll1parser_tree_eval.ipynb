```python
class S:
    def __init__(self, tchild, uchild): # self : 자기자신 가리키는 레퍼런스 
        self.t = tchild
        self.u = uchild
    def __repr__(self):
        return f'{self.t}{self.u}'
    def evaluate(self):        
        return self.u.evaluate(self.t.evaluate())
        
class T:
    def __init__(self, fchild, gchild):
        self.f = fchild
        self.g = gchild
    def __repr__(self):
        return f'{self.f}{self.g}'
    def evaluate(self):
        return self.g.evaluate(self.f.evaluate())
         
class U:
    def __init__(self, op=None, tchild=None, uchild=None):
        self.t = tchild
        self.u = uchild
        self.op = op
    def __repr__(self):
        
        if not self.t and not self.u and not self.op:
            return ''
        elif not self.t and not self.op:
            return f'{self.u}'
        elif not self.u and not self.op:
            return f'{self.t}'
        elif self.t and self.u and not self.op:
            return f'{self.t}{self.u}'
        elif self.op and self.t and not self.u:
            return f'{self.op}{self.t}'
        elif self.op and self.u and not self.t:
            return f'{self.op}{self.u}'
        else:
            return f'{self.op}{self.t}{self.u}'
        
    def evaluate(self, tchild2): # t 의 결과
        if not self.op:
            return tchild2
        else:
            return tchild2 + self.u.evaluate(self.t.evaluate())


class F: # 클래스 나눠서 하는게 좋음, C에 대한 클래스 따로 만드는게 좋음
    def __init__(self, child1,sign=None,sign2=None):
        self.child1 = child1
        self.sign = sign
        self.sign2 = sign2
    def __repr__(self):
        if self.sign and not self.sign2:
            return f'{self.sign}{self.child1}'
        elif not self.sign and not self.sign2:
            if self.child1 == 'x':
                print(self.child1,'=', end='')
                self.child1 = float(input())
            return f'{self.child1}'
        elif self.sign and self.child1 and self.sign2:
            return f'{self.sign}{self.child1}{self.sign2}'
    def evaluate(self):
        if self.sign and not self.sign2: # -F
            return -float(str(self.child1))
        elif not self.sign: # C
            return float(str(self.child1))
        elif self.sign and self.sign2: # (S)
            s = S(self.child1.t, self.child1.u)
            return s.evaluate()
    
        
        
class G:
    def __init__(self, op=None, fchild=None, gchild=None):
        self.op = op
        self.f = fchild
        self.g = gchild
    def __repr__(self):
        if not self.f and not self.g and not self.op:
            return ''
        elif not self.f and not self.op:
            return f'{self.g}'
        elif not self.g and not self.op:
            return f'{self.f}'
        elif self.f and self.g and not self.op:
            return f'{self.f}{self.g}'
        elif self.f and self.g and self.op:
            return f'{self.op}{self.f}{self.g}'  
        
    def evaluate(self, fchild2):
        if not self.op:
            return fchild2 ##### 이거 잘 모르겠음 입실론 표현!!
        else:
            print(fchild2, self.f.evaluate(), self.g.evaluate(self.f.evaluate()))
            print('fchild2 * self.g.evaluate(self.f.evaluate())=',fchild2 * self.g.evaluate(self.f.evaluate()))
            return fchild2 * self.g.evaluate(self.f.evaluate())
```

```python
def takeS(tokens): 
    t = takeT(tokens) 
    u = takeU(tokens)
    return S(t, u)
    
def takeT(tokens):
    f = takeF(tokens)
    g = takeG(tokens)
    return T(f, g)
    
def takeU(tokens):
    if len(tokens)>0 and tokens[0] == '+':
        op = takeIt('+',tokens)
        t = takeT(tokens)
        u = takeU(tokens)
        return U('+', t, u)
    else: 
        return U(None,None,None) 

def takeIt(tokenType, tokens):
    if len(tokens)>0 and tokens[0]== tokenType:
        op = tokens.pop(0)
        return op
    
def takeF(tokens): 
    if len(tokens)>0 and tokens[0] == '-':
        op = takeIt('-', tokens)
        f = takeF(tokens)
        return F(f, op, None)
    
    elif len(tokens)>0 and tokens[0] == '(':
        op = takeIt('(',tokens)
        s = takeS(tokens)
        op2 = takeIt(')',tokens)
        return F(s, op, op2)
    
    else: # C
        num = takeIt(tokens[0],tokens)
        return F(num,None,None) #numF(num)

def takeG(tokens):
    if len(tokens)>0 and tokens[0] == '*':
        op = takeIt('*',tokens)
        f = takeF(tokens)
        g = takeG(tokens)
        return G(op, f, g)
    else:
        return G(None,None,None)
    
def tokenize(expStr):
    import re
    RE = re.compile(r'(?:(?<=[^\d\.])(?=\d)|(?=[^\d\.]))', re.MULTILINE)
    result = [x for x in re.sub(RE, " ", expStr).split(" ") if x]
     
    return result
```
